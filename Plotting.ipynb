```julia
D = load("TimesOptDirUni.jld2")
```

```julia
T = D["My"]
errorbars = D["My error"]
T2 = D["3"]
errorbars2 = D["3 Error"];
```

```julia
N = 9
xlab = "Number of nodes"
ylab = "Time in seconds"
x = collect(100:100:N*100)  
plot(x,T[1:N],xlabel=xlab,ylabel=ylab, yerr = [errorbars[1:N,:];errorbars[1:N,1]],label="Our Algorithm",legend=:topleft)
plot!(x,T2[1:N],xlabel=xlab,ylabel=ylab, yerr = [errorbars2[1:N,:];errorbars2[1:N,1]], label="Brickell et al.")
```
